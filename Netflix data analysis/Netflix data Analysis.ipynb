# Analyzing Content on Netflix :

```python
df1 = df[['type', 'release_year']]
df1
```

```python
df1 = df1.rename(columns={'release_year':'Release_year'})
```

```python
df2 = df1.groupby(['Release_year','type']).size().reset_index(name='Total Content')
```

```python
df2 = df2[df2['Release_year']>=2010]
df2
```

```python
fig3 = px.line(df2, x='Release_year', y='Total Content', color='type', title='Trend of content produced over the years on Netflix')
fig3.show()
```

The above line graph shows that there has been a decline in the production of the content for both movies and other shows since 2018.

# **At last, to conclude our analysis, I will analyze the sentiment of content on Netflix:**

```python
dfx = df[['release_year', 'description']]
dfx
```

```python
dfx = dfx.rename(columns={'release_year':'Release_year'})

for index, row in dfx.iterrows():
  z = row['description']
  testimonial = TextBlob(z)
  p = testimonial.sentiment.polarity

  if p==0:
    sent = 'Neural'
  elif p>0:
    sent='Positive'
  else: 
    sent = 'Negative'

  dfx.loc[[index,2], 'sentiment'] = sent

dfx = dfx.groupby(['Release_year', 'sentiment']).size().reset_index(name='Total Content')
dfx = dfx[dfx['Release_year']>=2010]
```

```python
fig4 = px.bar(dfx, x='Release_year', y='Total Content',color='sentiment', title='Sentiment of content on Netflix')
fig4.show()
```

## So the above graph shows that the overall positive content is always greater than the neutral and negative content combined.
